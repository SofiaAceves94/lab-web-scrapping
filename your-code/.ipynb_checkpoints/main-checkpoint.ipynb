#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url3 = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
html3 = requests.get(url3).content
soup3 = BeautifulSoup(html3, 'lxml')
images = soup3.select('td a[class="image"]')
for ima in images:
    if ima.has_attr('href'):
        print(ima.attrs['href'])
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html7 = requests.get(url7).content
soup7 = BeautifulSoup(html7, 'lxml')
#data = soup7.select('table')
data_dh = soup7.select("b")[:20]
#raw_data = [dat.text for dat in data]
date = [re.findall('\d{4}-\d{2}-\d{2}',dat.text) for dat in data_dh]
time = [re.findall('\d{2}:\d{2}',dat.text) for dat in data_dh]
#clean_data_d = re.findall('\d{4}-\d{2}-\d{2}',raw_data_dh)
#clean_data_h = re.findall('\d{3}:\d{2}:\d{2}',raw_data_dh)
data_latlon_num = soup7.select("td.tabev1")[:40]
latlon_num = [re.findall(r'\d+\.\d+',dat.text) for dat in data_latlon_num]
#clean_data_latlon_num = re.findall(r'\d+\.\d+',raw_data_latlon_n)
data_latlon_dir = soup7.select("td.tabev2")[:60]
latlon_dir = [re.findall(r'[A-Z]', dat.text) for dat in data_latlon_dir if re.findall(r'[A-Z]', dat.text) != []]
data_latlon = list(zip(latlon_num,latlon_dir))
#    for erre in rawdat2:
#        r + erre
data_region = soup7.select("td.tb_region")[:20]
region = [dat.text.replace('\xa0','') for dat in data_region]
#clean_data_reg = re.findall(r'[A-Z]+|[A-Z]+\s+[A-Z]+|',raw_data_reg) - LO MÁS CERCANO A LO QUE LLEGUÉ EN REGEX, PERO NO LO LOGRÉ DEL TODO

#CÓMO JUNTARLO PARA HACERLO EN PANDAS?
# = pd.DataFrame(data[4:], columns=colnames)
lon = data_latlon[1::2]
lat = data_latlon[0::2]

earthquake_df= pd.DataFrame(list(zip(date,time,lat,lon,region)),columns=['Date','Time','Latitude','Longitude','Region Name'])
earthquake_df
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
#sólo obtengan fecha, lugar y nombres de lo próximos y pasados eventos
url8 ='https://mlh.io/seasons/2021/events'
````

````python
#your code
html8 = requests.get(url8).content
soup8 = BeautifulSoup(html8, 'lxml')
data_name = soup8.select('h3.event-name')
name = [dat.text for dat in data_name]
data_date = soup8.select('p.event-date')
date = [dat.text for dat in data_date]
data_loc = soup8.select('div.event-location')
loc = [dat.text.strip("\n").replace("\n          "," ") for dat in data_loc]

hackaton_df = pd.DataFrame(list(zip(name,date,loc)),columns=['Name','Date','Location'])
hackaton_df.head(25)
````
